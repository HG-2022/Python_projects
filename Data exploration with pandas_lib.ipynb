```python
import pandas as pd

url= 'https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv'
df= pd.read_csv(url)
print(df.head(5))
df
```

```python

#### printing top and bottom of data set
df.head(10)
df.tail(10)


### Reading headers

print(df.columns)

df.columns ## also prints col

## printing specific collumns
print(df[['Name','Type 1', 'HP']])
df[['Name','HP']]
print(df.head(5))## printing rows 

#### iloc function
## to return specific row, iloc (integer location) is used
print(df.iloc[4]) 
print(df.iloc[0:4])

## to print value in a specific row and column by using their respective numbers instead of names
print(df.iloc[3,5])


### iterrows()
## iterate rows using iterrows(). iterrows is for loop in python
for index, row in df.iterrows():
  #print(index,row)
  print(index,row['Type 1'])
  
  ### Unique function
  ### dataframe[column name].unique function
  print(df['Name'].unique())
df['Name'].unique() ## also prints
 

df.columns


### df.loc()
## selecting rows based on a specific condition . similar to where condtion in sql
df.loc[df['Type 1']== "Grass"]

df.describe()
10+20

# checks is there any missing values
df.isna().any()

#checks is there any null values. true means null. false means no null
df.isnull() 

## describes shape rows and columns of a table 
df.shape
rows,columns = df.shape
rows
columns

##min max mean std of a column

df['Attack'].max()
df['minimum']=df['Attack'].min()
df['Defense'].std()
df.head(5)

#conditional queries
df[df.Attack >=50]

## return row where attack is maximum
df[df['Attack']== df['Attack'].max()]

## i dont want entire row. give me just Name column where Attack is max
df.Name[df['Attack']== df['Attack'].max()]

###I dont want entire row. I want more than 1 columns from the row where defense is max 
df[['Name','HP','Defense']][df['Defense']==df['Defense'].max()]


## set index
df.set_index('Type 1', inplace=True)
df
df.loc['Fire']

#to reset index.
df.reset_index(inplace=True)
df


## making changes to data
df.head(5) 
df['total']=df['Attack']+df['HP']
49+45

#drop collumn
df=df.drop['total']
df
df['sumofcolumns']=df.iloc[:,4:8].sum(axis=1) ## axis= 1 sums horizontally
df.head(5)

##axis= 0 sums vertically
df['sum_horizontal']=df.iloc[:,4].sum(axis=0)
df

## to save as csv
df.to_csv('modified.csv', index=False) ## to remove auto index say index=False
df.to_csv('tabseperated',index=False,sep='\t')
df.to_excel('modifiedtoexcel.xlsx',index=False)

### fitering data
new_df = df.loc[(df['Type 1'] == 'Grass') | (df['Type 2']== 'Poison') & (df['HP']>65)]
new_df
new_df.to_csv('fitereddata.csv',index=True)
new_df.reset_index(drop = True, inplace= True)
new_df
df
df.loc[df['Name'].str.contains('Mega')]

## using str.contains to filter

ndf=df.loc[df['Name'].str.contains('Mega')]
ndf.reset_index(drop = True, inplace= True)
ndf
df.Name.str.upper()

df[~df['Name'].str.contains('Mega')] ## ~ indicates no


#### filtering with regex
import re
filt=df['Name'].str.contains('^S[a-z]*',flags=re.I,regex=True) ### ^S[a-z]*     ^indicating starting with S[anything a-z] *repeat anytimes 0 to any #of times 
#filt=df['Name'].str.contains('Mega')
df.loc[filt,'Speed']  ## df returns rows from filt and speed collumn



##### edit/ updating data with 2 conditions. updating legendary column where condition 1 and condition 2 are met.
df.loc[(df['Type 2']== 'Poison') & (df['Name']== 'Bulbasaur'), 'Legendary']='True'   
df.to_csv('updated.csv', index=True)   

df.loc[df['Type 2']=='greengrass','Type 2']='Poison'
df


## to edit 2 columns at the same time with given condition
df.loc[df['Speed'] > 50, ['Name', 'Legendary']]= ['superfast_pokemon','idk']
df



## groupby with single group and singlle coumn
df.groupby(['Type 1']) ['Defense'].max().sort_values('Defense',ascending=False)                                                                      


##groupby with multiple groups. as_index sets index for groups. to arrange values ascending/desecnding use function sort_values('coll name to arrange',ascending=False)
df.groupby(['Type 1', 'Type 2']) ['Defense'].max()#.sort_values('Defense',ascending=False)
df.loc[df['Type 2']=='Bug','Defense']
df.loc[df['Type 2']=='Bug']





```
